 **Visualize Channel Performance**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Create boxplots for numerical fields
plt.figure(figsize=(15, 6))

# Boxplot for Impressions
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Impressions'])
plt.title('Impressions')

# Boxplot for Clicks
plt.subplot(1, 3, 2)
sns.boxplot(y=df['Clicks'])
plt.title('Clicks')

# Boxplot for Acquisition_Cost
plt.subplot(1, 3, 3)
sns.boxplot(y=df['Acquisition_Cost'])
plt.title('Acquisition Cost')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore

# Calculate Z-Scores for numerical fields
df['Impressions_ZScore'] = zscore(df['Impressions'])
df['Clicks_ZScore'] = zscore(df['Clicks'])
df['Acquisition_Cost_ZScore'] = zscore(df['Acquisition_Cost'])

# Identify outliers (Z-Score > 3 or < -3)
outliers_impressions = df[(df['Impressions_ZScore'] > 3) | (df['Impressions_ZScore'] < -3)]
outliers_clicks = df[(df['Clicks_ZScore'] > 3) | (df['Clicks_ZScore'] < -3)]
outliers_cost = df[(df['Acquisition_Cost_ZScore'] > 3) | (df['Acquisition_Cost_ZScore'] < -3)]

# Display outliers
print("Outliers in Impressions:")
print(outliers_impressions[['Campaign_ID', 'Impressions', 'Impressions_ZScore']])

print("\nOutliers in Clicks:")
print(outliers_clicks[['Campaign_ID', 'Clicks', 'Clicks_ZScore']])

print("\nOutliers in Acquisition Cost:")
print(outliers_cost[['Campaign_ID', 'Acquisition_Cost', 'Acquisition_Cost_ZScore']])
```

```python
# Calculate IQR for numerical fields
Q1_impressions = df['Impressions'].quantile(0.25)
Q3_impressions = df['Impressions'].quantile(0.75)
IQR_impressions = Q3_impressions - Q1_impressions

Q1_clicks = df['Clicks'].quantile(0.25)
Q3_clicks = df['Clicks'].quantile(0.75)
IQR_clicks = Q3_clicks - Q1_clicks

Q1_cost = df['Acquisition_Cost'].quantile(0.25)
Q3_cost = df['Acquisition_Cost'].quantile(0.75)
IQR_cost = Q3_cost - Q1_cost

# Identify outliers
outliers_impressions_iqr = df[(df['Impressions'] < (Q1_impressions - 1.5 * IQR_impressions)) |
                              (df['Impressions'] > (Q3_impressions + 1.5 * IQR_impressions))]

outliers_clicks_iqr = df[(df['Clicks'] < (Q1_clicks - 1.5 * IQR_clicks)) |
                         (df['Clicks'] > (Q3_clicks + 1.5 * IQR_clicks))]

outliers_cost_iqr = df[(df['Acquisition_Cost'] < (Q1_cost - 1.5 * IQR_cost)) |
                       (df['Acquisition_Cost'] > (Q3_cost + 1.5 * IQR_cost))]

# Display outliers
print("Outliers in Impressions (IQR Method):")
print(outliers_impressions_iqr[['Campaign_ID', 'Impressions']])

print("\nOutliers in Clicks (IQR Method):")
print(outliers_clicks_iqr[['Campaign_ID', 'Clicks']])

print("\nOutliers in Acquisition Cost (IQR Method):")
print(outliers_cost_iqr[['Campaign_ID', 'Acquisition_Cost']])
```

```python
# Remove outliers using IQR method
df_cleaned = df[~((df['Impressions'] < (Q1_impressions - 1.5 * IQR_impressions)) |
                  (df['Impressions'] > (Q3_impressions + 1.5 * IQR_impressions)))]
df_cleaned = df_cleaned[~((df['Clicks'] < (Q1_clicks - 1.5 * IQR_clicks)) |
                          (df['Clicks'] > (Q3_clicks + 1.5 * IQR_clicks)))]
df_cleaned = df_cleaned[~((df['Acquisition_Cost'] < (Q1_cost - 1.5 * IQR_cost)) |
                          (df['Acquisition_Cost'] > (Q3_cost + 1.5 * IQR_cost)))]

# Check the size of the cleaned dataset
print("Original Dataset Size:", df.shape)
print("Cleaned Dataset Size:", df_cleaned.shape)
```

```python
# Cap outliers using IQR method
df['Impressions'] = df['Impressions'].clip(lower=Q1_impressions - 1.5 * IQR_impressions,
                                           upper=Q3_impressions + 1.5 * IQR_impressions)
df['Clicks'] = df['Clicks'].clip(lower=Q1_clicks - 1.5 * IQR_clicks,
                                 upper=Q3_clicks + 1.5 * IQR_clicks)
df['Acquisition_Cost'] = df['Acquisition_Cost'].clip(lower=Q1_cost - 1.5 * IQR_cost,
                                                     upper=Q3_cost + 1.5 * IQR_cost)

# Check the capped values
df[['Impressions', 'Clicks', 'Acquisition_Cost']].describe()
```

```python
# Apply log transformation to reduce the impact of outliers
import numpy as np

df['Impressions_Log'] = np.log1p(df['Impressions'])
df['Clicks_Log'] = np.log1p(df['Clicks'])
df['Acquisition_Cost_Log'] = np.log1p(df['Acquisition_Cost'])

# Check the transformed values
df[['Impressions_Log', 'Clicks_Log', 'Acquisition_Cost_Log']].describe()
```
